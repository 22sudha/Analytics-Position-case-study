```python
#load the cleaned User Gameplay data by skipping initial rows
user_df=pd.read_csv('/content/User Gameplay data.csv',skiprows=3)
user_df.head()
```

```python
#convert 'Datetime' column to datetime format
user_df['Datetime']=pd.to_datetime(user_df['Datetime'],format='%d-%m-%Y %H:%M')
user_df.info()
```

```python
#load deposit data(skip top 3 junk rows)
deposit_df=pd.read_csv('/content/Deposit Data.csv',skiprows=3)
deposit_df.head()
```

```python
#convert 'Datetime' to datetime format in deposit_df
deposit_df['Datetime']=pd.to_datetime(deposit_df['Datetime'],format='%d-%m-%Y %H:%M')
deposit_df.info()
```

```python
#load Withdrawal data(skip top 3 junk rows)
withdrawal_df=pd.read_csv('/content/Withdrawal Data.csv',skiprows=3)
withdrawal_df.head()
```

```python
#Convert 'Datetime' to datetime in withdrawal_df
withdrawal_df['Datetime']=pd.to_datetime(withdrawal_df['Datetime'],format='%d-%m-%Y %H:%M')
withdrawal_df.info()
```

```python
def assign_slot(dt):
    if dt.hour < 12:
        return'S1'
    else:
        return'S2'
user_df['Slot']=user_df['Datetime'].apply(assign_slot)
deposit_df['Slot']=deposit_df['Datetime'].apply(assign_slot)
withdrawal_df['Slot']=withdrawal_df['Datetime'].apply(assign_slot)
```

```python
#data slots which is mention in the assigment
filter_conditions=[
    ('2022-10-02','S1'),
    ('2022-10-16','S2'),
    ('2022-10-18','S1'),
    ('2022-10-26','S2')
]
from datetime import datetime
#convert string dates to datetime.date for filtering
dates_slots=[(datetime.strptime(date_str,'%Y-%m-%d').date(),slot) for date_str, slot in filter_conditions]
#function to filter data by date and slot
def filter_data_by_date_slot(df, date_col='Datetime'):
    filtered={}
    for date_val,slot_val in dates_slots:
        temp_df=df[(df[date_col].dt.date==date_val)&(df['Slot']==slot_val)]
        filtered[(date_val, slot_val)]=temp_df
    return filtered
#filter each dataset
user_filtered=filter_data_by_date_slot(user_df)
deposit_filtered=filter_data_by_date_slot(deposit_df)
withdrawal_filtered=filter_data_by_date_slot(withdrawal_df)
#Show number of records for each filtered subset as a check
for key in user_filtered:
    print(f"User Data on {key[0]}Slot{key[1]}:{len(user_filtered[key])} rows")
for key in deposit_filtered:
    print(f"Deposit Data on {key[0]}Slot{key[1]}:{len(deposit_filtered[key])} rows")
for key in withdrawal_filtered:
    print(f"Withdrawal Data on {key[0]}Slot{key[1]}:{len(withdrawal_filtered[key])} rows")
```

```python
def calculate_loyalty_points(user_df, deposit_df, withdrawal_df):
    games_played=user_df.groupby('User ID')['Games Played'].sum()
    total_deposit_amount=deposit_df.groupby('User Id')['Amount'].sum()
    total_withdrawal_amount=withdrawal_df.groupby('User Id')['Amount'].sum()
    deposit_counts=deposit_df.groupby('User Id').size()
    withdrawal_counts=withdrawal_df.groupby('User Id').size()
    #dataframe
    df=pd.DataFrame({'Games Played': games_played})
    df=df.join(total_deposit_amount.rename('Total Deposit Amount'))
    df=df.join(total_withdrawal_amount.rename('Total Withdrawal Amount'))
    df=df.join(deposit_counts.rename('Deposit Count'))
    df=df.join(withdrawal_counts.rename('Withdrawal Count'))
    #convert Na with 0
    df=df.fillna(0)
    #calculate loyalty points
    df['Points from Deposit']=0.01*df['Total Deposit Amount']
    df['Points from Withdrawal']=0.005*df['Total Withdrawal Amount']
    df['Points from Deposit-Withdrawal Count']=0.001*(df['Deposit Count']- df['Withdrawal Count']).clip(lower=0)
    df['Points from Games']= 0.2*df['Games Played']
    #final loyalty Points
    df['Loyalty Points']=df['Points from Deposit']+df['Points from Withdrawal']+\
                          df['Points from Deposit-Withdrawal Count'] + df['Points from Games']
    #reset index for cleaner output
    df=df.reset_index().rename(columns={'User ID': 'User Id'})
    return df[['User Id', 'Loyalty Points', 'Games Played']]
#calculate all four filtered date slots and store results
loyalty_points_results={}
for key in dates_slots:
    u=user_filtered[key]
    d=deposit_filtered[key]
    w=withdrawal_filtered[key]
    loyalty_df=calculate_loyalty_points(u, d, w)
    loyalty_points_results[key]=loyalty_df.sort_values(by='Loyalty Points',ascending=False)
#show results for 2nd October S1for ex:
loyalty_points_results[(datetime.strptime('2022-10-02','%Y-%m-%d').date(),'S1')].head()
```

```python
#filter entire october month data for all datasets
user_oct=user_df[(user_df['Datetime'].dt.month==10)&(user_df['Datetime'].dt.year==2022)]
deposit_oct=deposit_df[(deposit_df['Datetime'].dt.month==10)&(deposit_df['Datetime'].dt.year==2022)]
withdrawal_oct=withdrawal_df[(withdrawal_df['Datetime'].dt.month==10)&(withdrawal_df['Datetime'].dt.year==2022)]
#calculate loyalty points for the whole month
loyalty_oct=calculate_loyalty_points(user_oct, deposit_oct, withdrawal_oct)
#rank players by loyalty points and games played (descending order)
loyalty_oct['Rank']=loyalty_oct.sort_values(['Loyalty Points','Games Played'],ascending=[False, False])\
                                .reset_index().index+1
#mention according to rank
loyalty_oct=loyalty_oct.sort_values('Rank')
#show top 10 players
loyalty_oct.head(10)
```

```python
avg_deposit_amount=deposit_oct['Amount'].mean()
total_deposit_per_user=deposit_oct.groupby('User Id')['Amount'].sum()
avg_deposit_per_user=total_deposit_per_user.mean()
total_games_per_user=user_oct.groupby('User ID')['Games Played'].sum()
avg_games_per_user=total_games_per_user.mean()
#printing the records
print(f"Average deposit amount (per transaction):{avg_deposit_amount:.2f}")
print(f"Average deposit amount per user (total deposits per user averaged):{avg_deposit_per_user:.2f}")
print(f"Average number of games played per user:{avg_games_per_user:.2f}")
```
